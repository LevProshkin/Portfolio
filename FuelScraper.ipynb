```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.options import Options
from datetime import datetime, timedelta

# Path to your EdgeDriver (msedgedriver.exe)
edge_driver_path = 'C:/Drivers/msedgedriver.exe'  # Replace with your actual path to msedgedriver.exe

# Create a Service object for EdgeDriver
service = Service(executable_path=edge_driver_path)

# Create options for Edge
options = Options()
options.add_argument('--start-maximized')  # Optional: Start Edge maximized

# Initialize a WebDriver for Microsoft Edge
driver = webdriver.Edge(service=service, options=options)



# Base URLs
base_urls = [
    'https://index.minfin.com.ua/ua/markets/fuel/tm/klo/',
    'https://index.minfin.com.ua/ua/markets/fuel/tm/socar/',
    'https://index.minfin.com.ua/ua/markets/fuel/tm/wog/',
    'https://index.minfin.com.ua/ua/markets/fuel/tm/brsmnafta/',
    'https://index.minfin.com.ua/ua/markets/fuel/tm/okko/'
]

# Date range
start_date = datetime(2020, 1, 1)
end_date = datetime(2024, 9, 1)

# Generate all months from start_date to end_date
dates = []
current_date = start_date
while current_date <= end_date:
    dates.append(current_date.strftime('%Y-%m'))  # Format as YYYY-MM
    current_date += timedelta(days=32)  # Adding more than 30 days to ensure jumping to the next month
    current_date = current_date.replace(day=1)  # Set day to 1 to avoid skipping months

# Generate URLs for each base URL and each date
generated_urls = []
for base_url in base_urls:
    for date in dates:
        generated_urls.append(base_url + date + '/')

# Initialize an empty list to store the data from all pages
all_data = []

# Loop through each URL
for url in generated_urls:
    # Open the URL
    driver.get(url)

    # Wait for the page to load
    time.sleep(5)

    # Find the article with id 'idx-content' where the tables are located
    article = driver.find_element(By.ID, "idx-content")

    # Scrape the company name from the <h1> tag
    company_name = article.find_element(By.TAG_NAME, 'h1').text.strip().split('«')[-1].split('»')[0]

    # Find the table inside the article
    table = article.find_element(By.CSS_SELECTOR, 'table.zebra')

    # Find the headers of the table (column names)
    headers = table.find_elements(By.TAG_NAME, 'th')
    header_names = [header.text.strip() for header in headers]
    
    # Find the rows in the table body
    rows = table.find_elements(By.TAG_NAME, 'tr')

    # Loop through the rows and extract the data
    for row in rows[1:]:  # Skip the header row
        columns = row.find_elements(By.TAG_NAME, 'td')
        row_data = {}

        # Loop through columns and assign the values to corresponding headers
        for i, column in enumerate(columns):
            if i < len(header_names):
                header = header_names[i]
                value = column.text.strip()

                # Add data to row_data with header as the key
                row_data[header] = value

        # Add the company name to the row data
        row_data['Company Name'] = company_name

        # Append the row data to all_data list
        all_data.append(row_data)

# Create a DataFrame from the collected data from all URLs
df = pd.DataFrame(all_data)

# Display the DataFrame
print(df)

# Close the browser
driver.quit()
```
